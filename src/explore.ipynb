```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_esc = scaler.fit_transform(x_train,y_train)
x_test_esc = scaler.transform(x_test)

x_train_esc = pd.DataFrame(x_train_esc,index = x_train.index,columns=x_train.columns)
x_test_esc = pd.DataFrame(x_test_esc,index = x_test.index,columns=x_test.columns)
```

```python
from sklearn.linear_model import LinearRegression, Lasso

modelRL = LinearRegression()
modelLasso = Lasso()

modelRL.fit(x_train_esc,y_train)
modelLasso.fit(x_train_esc,y_train)

```

```python
y_pred_RL = modelRL.predict(x_test_esc)
y_pred_Lasso = modelLasso.predict(x_test_esc)

from sklearn.metrics import r2_score, root_mean_squared_error

print(f'Para la Regresion lineal se obtuvo un RMSE de:{root_mean_squared_error(y_test,y_pred_RL)} y un R2 de:{r2_score(y_test,y_pred_RL)}')

print(f'Para la Regresion lineal se obtuvo un RMSE de:{root_mean_squared_error(y_test,y_pred_Lasso)} y un R2 de:{r2_score(y_test,y_pred_Lasso)}')
```

```python
alphas = np.arange(0.1, 20.5, 0.5) 
r2_scores = []

print("Entrenando modelos Lasso con distintos alphas...")

for alpha in alphas:
    model = Lasso(alpha=alpha, max_iter=10000)
    model.fit(x_train_esc, y_train)
    y_pred = model.predict(x_test_esc)
    
    score = r2_score(y_test, y_pred)
    r2_scores.append(score)

plt.figure(figsize=(10, 6))
plt.plot(alphas, r2_scores, marker='o', linestyle='-', color='b')
plt.title("Evolucin de R2 en funcin de Alpha (Lasso)")
plt.xlabel("Alpha (Hiperparmetro de Regularizacin)")
plt.ylabel("R2 Score")
plt.grid(True)
plt.show()

best_alpha_manual = alphas[np.argmax(r2_scores)]
best_r2_manual = max(r2_scores)
print(f"El mejor alpha en el rango analizado es: {best_alpha_manual} con un R2 de: {best_r2_manual:.4f}")
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'alpha': [0.01, 0.1, 1, 5, 10, 20, 50, 100], 
    'max_iter': [1000, 5000, 10000]              
}

grid_search = GridSearchCV(
    estimator=Lasso(),
    param_grid=param_grid,
    cv=5,             
    scoring='r2',     
    n_jobs=-1,        
    verbose=1
)

print("Iniciando optimizacin con GridSearchCV...")
grid_search.fit(x_train_esc, y_train)

print("\n--- Mejores Hiperparmetros Encontrados ---")
print(f"Mejores Parmetros: {grid_search.best_params_}")
print(f"Mejor Score R2 (CV): {grid_search.best_score_:.4f}")
```
